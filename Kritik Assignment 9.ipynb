```python
import numpy as np #import libraries 
import sympy as sp
x, y = sp.symbols('x y') #define x and y as symbols
```

```python
f = (sp.exp(x))*sp.sin(y)+(y**3) #defining function f for part a
df_dx = sp.diff(f, x) #differentiate f with respect to x
df_dy = sp.diff(f, y) #differentiate f with respect to y

print (df_dx) #print the resulting formulas
print (df_dy)
```

```python
g = (x**2)*y+x*(y**2) #define fucntion g for part b

def gradient_g (g, x, y): #define the gradient vector as a function of the derivatives of g w.r.t x and y
  return (sp.diff(x), sp.diff(y)) 

print (gradient_g (g, 1, -1)) #evaluate at the given point (1, -1)
```

```python
import sympy as sp #import libraries
import numpy as np
import matplotlib.pyplot as plt

x, y = sp.symbols('x y') #define symbols
j = x**3 - 3*x*y + y**3 #define function j for part d
j_func = sp.lambdify((x, y), j, 'numpy') #convert to numerical 

x_vals = np.linspace(-3, 3, 400) #define x, y, and z values 
y_vals = np.linspace(-3, 3, 400)
X, Y = np.meshgrid(x_vals, y_vals)
Z = j_func(X, Y)

plt.contourf(X, Y, Z, levels=50, cmap='viridis') # set up plot
plt.colorbar()
plt.title('Contour plot of $j(x, y) = x^3 - 3xy + y^3$')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.show()
```

```python
x, y = sp.symbols('x y') #same steps as previous cell, just for part e with a different function
f = (x-y)/(1+(x**2)+(y**2))
f_func = sp.lambdify((x, y), f, 'numpy')

x_vals = np.linspace(-3, 3, 400)
y_vals = np.linspace(-3, 3, 400)
X, Y = np.meshgrid(x_vals, y_vals)
Z = f_func(X, Y)

plt.contourf(X, Y, Z, levels=50, cmap='viridis')
plt.colorbar()
plt.title('Contour plot of $f(x, y) = (x-y)/(1+x^2+y^2)$')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.show()
```
