```python
import matplotlib.pyplot as plt #import libraries 
import numpy as np

def gradient_descent(f, learning_rate, initial_point):
    def deriv(f, base_point): #this allows us to define a function that takes in a specific f (function) that starts at a specific base point
        return (f(base_point+10**(-5))-f(base_point))/(10**(-5)) 
    
    x_coords=[initial_point] #lists to store x and y coordinates 
    y_coords=[f(initial_point)]
    
    point = initial_point-learning_rate*deriv(f, initial_point) #Makes each new point closer to the minimum by decreasing its size by a learning rate multiplied by f'(x)
    for i in range (1000): #the chosen range is the number of times I want the code to run
        point = point-learning_rate*deriv(f, point) #this ensures that the function finds a new (closer) point each time the code runs
        x_coords.append(point) #adds each consecutive point to the x and y coordinate lists
        y_coords.append(f(point))
        if deriv(f, point)<0.00000001: #this stops the code from running when the slope of the derivative gets very close to zero, aka at/near x_n
            break 
            
    plot_range=np.linspace(min(x_coords)-0.5, max(x_coords)+0.5, 10000) #setting plotting range
    function_range = [f(i) for i in plot_range]
    plt.plot(plot_range, function_range) #plots f(x)
    plt.plot(x_coords, y_coords) #plots coordinate points 
    return round(x_coords[-1],3), round(y_coords[-1],3) #returns final x and y values rounded to 3 decimal places 
```

```python
f = lambda x: abs(x) #This cell introduces the function to be taken in
```

```python
gradient_descent(f, 0.013, 0.1) #input values that works best to determine minimum
```
